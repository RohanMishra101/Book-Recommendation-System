##Popularity Based Recommendation System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num-ratings'},inplace=True)
num_rating_df
```

```python
print(ratings_with_name.dtypes)
```

```python
print(ratings_with_name['Book-Rating'].unique())
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
```

```python
ratings_with_name = ratings_with_name.dropna(subset=['Book-Rating'])
```

```python
# Calculate average rating per book title
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the column to 'avg-rating'
avg_rating_df.rename(columns={'Book-Rating': 'avg-rating'}, inplace=True)

# Display the result
print(avg_rating_df)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num-ratings'] >= 100].sort_values('avg-rating', ascending=False).head(1000)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title').shape
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author_x','Image-URL-M_x','num-ratings','avg-rating']]
```

```python
popular_df
```

##Clobrative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
users_how_read = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(users_how_read)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index 
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse=True)[1:20]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.append(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.append(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.append(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data
```

```python
recommend('Message in a Bottle')
```
